```python
import numpy as np 
import pandas as pd

my_arr=np.array([[1,2,3],[4,5,6],[8,'' , '']])

rown=['a','b','c']
colnames=['one','two','three']

mydataframe=pd.DataFrame(my_arr,index=rown,columns=colnames)
print(mydataframe)
print('method1')
print(mydataframe['one'])
print('method2')
print(mydataframe.one)
```

```python
myarr=np.array([[1,2,3],[3,4,5]])

rwnames=['a','b']
colnames=['one','two','three']

mydataframes=pd.DataFrame(myarr, index=rwnames,columns=colnames)

print(mydataframes)
print('method 1') 
print(mydataframes['one']) 
print('method 2') 
print(mydataframes.one)
```
